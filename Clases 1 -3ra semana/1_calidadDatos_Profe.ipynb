```python
def completitud_datos(df):
    return df.isnull().sum().sort_values(ascending=False) / df.shape[0]
```

```python
completitud_datos(df=df)
```

### Ingenieria de Datos

#### Ratings

```python
from sklearn.impute import SimpleImputer
```

```python
imputador = SimpleImputer(strategy="mean")
```

```python
df[['Rating']]
```

```python
df.Rating
```

```python
df[['Rating']] = imputador.fit_transform(df[['Rating']])
```

```python
df[['Rating']] =df[['Rating']].astype(float)
```

```python
completitud_datos(df=df)
```

#### Current Ver 

```python
imputador_cv = SimpleImputer(strategy='most_frequent')

df['Current Ver'] = df['Current Ver'].map(lambda x: str(x).strip().lower())

df['Current Ver']  = pd.Series(imputador_cv.fit_transform(df[['Current Ver']]).ravel(), index=df.index)

completitud_datos(df=df)
```

```python
df.describe()
```

#### Type

```python
df["Type"] = df["Type"].replace(np.NaN, ("Free"))
df["Type"] = df["Type"].astype(str)
```

```python
completitud_datos(df)
```
